```python
arr1 = [1,2,3]
arr2 = [4,5,6]
arr3 = list(zip(arr1,arr2))
print(arr3)

for element in arr3:
    print(element)

for i,j in arr3:
    print(i,j)

for i,j in zip(arr1,arr2):
    print(i,j)
```

```python
arr3 = arr1+arr2
print(arr3)

#reverse - reverse inplace it doesn't return anything
arr3.reverse()
print(arr3)

arr3.sort()
print(arr3)

#reverse sorting
arr3.sort(reverse=True)
print(arr3)
```

```python
#lambda is anonymous function and we can call inline
result = ( lambda x : x + 10 ) ( 20 )
print(result)

#custom sort array
arr = ["bob", "alice", "jane", "doe"]
arr.sort()
print(arr)

# Custom sort (by length of string)
arr.sort(key= lambda x : len(x))
print(arr)
```

```python
# list comprehension - concise way to create lists
arr = [ i for i in range(5) if i > 2 ] 
print(arr)

# 2-D lists
arr = [[0] * 4 for i in range(4)]
print(arr)
```

```python
#strings are similar to arrays
word = 'hello world'

for letter in word:
    print(letter)

for index,letter in enumerate(word):
    print(index,letter)

sliced = word[6:]
print(sliced)

from_last = word[:-2]
print(from_last)


#IMPORTANT - ARRAYS ARE MUTABLE BUT STRINGS ARE IMMUTABLE
#word[5] = 'c'

word = word + 'from china' # THIS CREATES A NEW STRING
print(word)


#ascii values
ascii = ord(word[5])
print(ascii)

#we can combine things
words = ['my','name','is','antony','gonsalvez']
word = " ".join(words)
print(word)
words = word.split(" ")
print(words)
```

```python
from collections import deque

queue = deque()
queue.append(1)
queue.append(2)
queue.appendleft(3)
print(queue)
queue.pop()
print(queue)
queue.appendleft(4)
queue.append(5)
queue.popleft()
print(queue)
```

# Hash Set

```python
hash_set = set()
hash_set.add(1)
hash_set.add(23)
hash_set.add(7)
hash_set.add(32)
hash_set.add(32)
print(hash_set)

# to check if something is in set
if 7 in hash_set:
    print("yes")
else:
    print("no")

if 5 in hash_set:
    print("yes")
else:
    print("no")


for item in hash_set:
    print(item)

#set comprehension similar to list comprehension - here it returns hash set
test = {i for i in hash_set}
print(test)
```

# Hash Map

```python
#hashm map is dictionary ie, key value pairs

hash_map = dict() # or {}
hash_map["shiy"] = 45
hash_map["l"] = 40
hash_map["n"] = 38
hash_map["shil"] = 35
print(hash_map)
print(len(hash_map))

print("shiy" in hash_map)
hash_map.pop("shiy")
print("shiy" in hash_map)


#dictionary comprehension
comp = { key:value for key,value in hash_map.items()}
print(comp)

for key in hash_map.keys():
    print(key)

for value in hash_map.values():
    print(value)

for key,value in hash_map.items():
    print(key,value)
```
