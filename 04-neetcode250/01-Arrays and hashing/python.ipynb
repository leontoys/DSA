```python
#if doesn't need paranthesis or curly braces
n = 1000
if n > 100 : 
    n = n - 1
elif n < 100:
    n = n + 1
else:
    n = n * n

print(n)
```

```python
#paranthesis needed for multi line conditions
a , b = 100, 200
if ( a > 100 and 
     b > 200 ):
    print('big numbers')
else:
    print('we can handle')
```

```python
# while
n = 5

print('count down begins')

while n >= 0 : 
    print(n)
    n = n - 1

print('go')
```

```python
print(math.floor(5/2))
print(math.ceil(5/2))
print(math.sqrt(16))
print(math.pow(5,2))
```

```python
#biggest numbers
#python numbers don't overflow and they are between -inf and +infy
max = float("inf")
min = float("-inf")

if max > 100:
    print(max)
if min < -100:
    print(min)

```

```python
#Python has lists that's the arrays
arr = [1,2,3]
print(arr)
```

```python
#it works like stack - ie, add to the top , remove from the top with o(1)
arr.append(4)
print(arr)
arr.pop()
print(arr)
```

```python
#we can insert at an index - this has time complexity of O(n)
arr.insert(1,9)
print(arr)
```

```python
#any item can be accessed at O(1) and we can reassign too
arr[1] = 0
print(arr)
```

```python
#array of length n
arr = [1] * 5
print(arr)
```

```python
# last value
val = arr[-1]
print(val)
val = arr[-2]
print(val)
val = arr[-3]
print(val)
```
