```python
#if doesn't need paranthesis or curly braces
n = 1000
if n > 100 : 
    n = n - 1
elif n < 100:
    n = n + 1
else:
    n = n * n

print(n)
```

```python
#paranthesis needed for multi line conditions
a , b = 100, 200
if ( a > 100 and 
     b > 200 ):
    print('big numbers')
else:
    print('we can handle')
```

```python
# while
n = 5

print('count down begins')

while n >= 0 : 
    print(n)
    n = n - 1

print('go')
```
