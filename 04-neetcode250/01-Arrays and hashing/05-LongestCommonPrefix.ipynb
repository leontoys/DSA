Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string "".

 

Example 1:

Input: strs = ["flower","flow","flight"]
Output: "fl"

Example 2:

Input: strs = ["dog","racecar","car"]
Output: ""
Explanation: There is no common prefix among the input strings.

 

Constraints:

    1 <= strs.length <= 200
    0 <= strs[i].length <= 200
    strs[i] consists of only lowercase English letters if it is non-empty.



```python
#                   then add it to the prefix
#               else
#                   stop

from typing import List

class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        prefix = ''
        # loop through each letter of first word
        for i in range(len(strs[0])):
            # loop from second word till last word
            for j in range(1,len(strs)):
                #if ith letter of first word with ith letter of jth word
                if i < len(strs[j]) - 1 and strs[0][i] == strs[j][i] and j == len(strs)-1:
                    #print(prefix)
                    prefix += strs[0][i]
                else:
                    #prefix = prefix[0:i]
                    break
        #print(prefix)
        return prefix

        
```

```python
sol = Solution()
strs = ["flower","flow","flight"]
print(sol.longestCommonPrefix(strs))
```

```python
strs = ["dog","racecar","car"]
print(sol.longestCommonPrefix(strs))
```
